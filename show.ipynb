```python
import matplotlib.pyplot as plt
import numpy as np
import os

area_name_list = ['western_pacific', 'eastern_pacific', 'Atlantic_Ocean', 'indian_ocean']
singular_list_vapor = np.load('singular_vapor.npy')
for i in range(4):
    plt.plot(np.square(singular_list_vapor[i]))
plt.legend(area_name_list)
plt.xlabel('kth')
plt.xticks(range(0,20,2))
plt.ylabel('singular_value')
plt.show()

#前n个奇异值的平方
np.square(singular_list_vapor)
```

```python
# 前n个奇异值的平方和
np.sum(np.square(singular_list_vapor), axis=1)
```

```python
singular_list_rain = np.load('singular_rain.npy')
for i in range(4):
    plt.plot(np.square(singular_list_rain[i]))
plt.legend(area_name_list)
plt.xlabel('kth')
plt.xticks(range(0,20,2))
plt.ylabel('singular_value_rain')
plt.show()
#前n个奇异值的平方
np.square(singular_list_vapor)
```

```python
# 前n个奇异值的平方和
np.sum(np.square(singular_list_vapor), axis=1)
```

```python
singular_all_vapor=np.load('singular_all_vapor.npy')
singular_all_rain=np.load('singular_all_rain.npy')
plt.plot(singular_all_rain)
plt.plot(singular_all_vapor)
plt.xlabel('kth')
plt.xticks(range(0,20,2))
plt.ylabel('singular_value_rain')
plt.legend(['rain','vapor'])
plt.show()
```
