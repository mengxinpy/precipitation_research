```python
import matplotlib.pyplot as plt
import numpy as np
import os

area_name_list = ['western_pacific', 'eastern_pacific', 'Atlantic_Ocean', 'indian_ocean']
singular_list_vapor = np.load('singular_vapor.npy')
for i in range(4):
    plt.plot(np.square(singular_list_vapor[i]))
plt.legend(area_name_list)
plt.xlabel('kth')
plt.ylabel('singular_value')
plt.show()

#前n个奇异值的平方
np.square(singular_list_vapor)
```

```python
# 前n个奇异值的平方和
np.sum(np.square(singular_list_vapor), axis=1)
```
