```python
import os
import matplotlib.pyplot as plt
from IPython.display import FileLink, FileLinks

# 创建一个文件夹来保存图像
output_dir = 'output_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 生成并保存多个图像
for i in range(5):
    plt.figure()
    plt.plot([0, 1, 2], [i, i**2, i**3])
    plt.title(f'Plot {i}')
    file_path = os.path.join(output_dir, f'plot_{i}.png')
    plt.savefig(file_path)
    plt.show()

# 在输出框中显示文件夹链接
display(FileLinks(output_dir))
```

```python
import os
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Image

# 创建一个文件夹来保存图像
output_dir = 'output_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 生成并保存多个图像
for i in range(5):
    plt.figure()
    plt.plot([0, 1, 2], [i, i**2, i**3])
    plt.title(f'Plot {i}')
    file_path = os.path.join(output_dir, f'plot_{i}.png')
    plt.savefig(file_path)
    plt.close()

# 创建一个交互式小部件来显示图像
def show_image(image_path):
    display(Image(filename=image_path))

image_files = sorted([os.path.join(output_dir, f) for f in os.listdir(output_dir) if f.endswith('.png')])
dropdown = widgets.Dropdown(options=image_files, description='Select Image:')
widgets.interactive(show_image, image_path=dropdown)
```

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure

# 生成并显示图像
fig = Figure()
canvas = FigureCanvas(fig)
ax = fig.add_subplot(111)
ax.plot([0, 1, 2], [0, 1, 4])
ax.set_title('Interactive Plot')
canvas.draw()

# 将图像嵌入到 Notebook 中
display(canvas.figure)
```
